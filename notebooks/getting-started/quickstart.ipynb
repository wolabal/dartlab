## Company 생성

모든 분석의 시작점이다. 종목코드 또는 회사명으로 생성한다.

```python
from dartlab import Company

c = Company("005930")
c.corpName, c.stockCode
```

```python
# 회사명으로도 생성 가능
kakao = Company("카카오")
kakao.corpName, kakao.stockCode
```

### 분기별 / 반기별

`fsSummary()` 메서드로 기간을 변경한다.

| 값 | 의미 | 포함 보고서 |
|-----|------|------------|
| `"y"` | 연간 (기본값) | 사업보고서 |
| `"q"` | 분기별 | 1분기 + 반기 + 3분기 + 사업 |
| `"h"` | 반기별 | 반기 + 사업 |

```python
quarterly = c.fsSummary(period="q")
quarterly.BS
```

```python
semiannual = c.fsSummary(period="h")
semiannual.IS
```

## 공시 텍스트

```python
# 사업의 내용 — 섹션별 분류
for s in c.business:
    print(f"[{s.key}] {s.title} ({s.chars:,}자)")
```

```python
# MD&A
c.mdna
```

```python
# 회사 개요
ov = c.overview
if ov:
    print(f"설립: {ov.founded}")
    print(f"소재지: {ov.address}")
    print(f"종속회사: {ov.subsidiaryCount}개")
```

## 전체 일괄 조회

`all()`을 호출하면 모든 모듈을 순회하면서 전체 데이터를 dict로 반환한다.

```python
d = c.all()
list(d.keys())
```

```python
d["BS"]
```

```python
d["dividend"]
```

## Result 객체 접근

property는 대표 DataFrame 하나를 반환한다. 모듈의 전체 데이터가 필요하면 `get()`을 사용한다.

```python
# property → 대표 DataFrame
c.audit
```

```python
# get() → 감사보수 (추가 데이터)
result = c.get("audit")
result.feeDf
```
