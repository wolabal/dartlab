## Company 생성

모든 분석의 시작점이다. 종목코드 또는 회사명으로 생성한다.

```python
from dartlab import Company

# 종목코드로 생성
c = Company("005930")
print(f"회사명: {c.corpName}")
print(f"종목코드: {c.stockCode}")

# 회사명으로도 생성 가능
kakao = Company("카카오")
print(f"\n회사명: {kakao.corpName}")
print(f"종목코드: {kakao.stockCode}")
```

## 재무제표

property 하나로 바로 DataFrame을 받는다. 괄호 없이 속성처럼 접근한다.

```python
print("=== 재무상태표 ===")
print(c.BS)
print("\n=== 손익계산서 ===")
print(c.IS)
print("\n=== 현금흐름표 ===")
print(c.CF)
```

### 분기별 / 반기별

`fsSummary()` 메서드로 기간을 변경한다. Bridge Matching으로 계정명 변경을 자동 추적한다.

| 값 | 의미 | 포함 보고서 |
|-----|------|------------|
| `"y"` | 연간 (기본값) | 사업보고서 |
| `"q"` | 분기별 | 1분기 + 반기 + 3분기 + 사업 |
| `"h"` | 반기별 | 반기 + 사업 |

```python
# 분기별
quarterly = c.fsSummary(period="q")
print("=== 분기별 재무상태표 ===")
print(quarterly.BS)

# 반기별
semiannual = c.fsSummary(period="h")
print("\n=== 반기별 손익계산서 ===")
print(semiannual.IS)
```

## 정기보고서

배당, 직원, 최대주주 등 정기보고서 데이터도 property로 바로 접근한다. 모든 property는 lazy loading + caching이다.

```python
# 배당
print("=== 배당 ===")
print(c.dividend)

# 직원 현황
print("\n=== 직원 현황 ===")
print(c.employee)

# 최대주주
print("\n=== 최대주주 ===")
print(c.majorHolder)
```

```python
# 감사의견
print("=== 감사의견 ===")
print(c.audit)

# 임원 현황
print("\n=== 임원 현황 ===")
print(c.executive)

# 임원 보수
print("\n=== 임원 보수 ===")
print(c.executivePay)
```

## K-IFRS 주석 (Notes)

재무제표 주석 데이터는 `c.notes`로 통합 접근한다. 12개 항목을 지원하며, 영문 속성과 한글 키 모두 사용할 수 있다.

```python
# 영문 속성
print("=== 재고자산 ===")
print(c.notes.inventory)

# 한글 키
print("\n=== 차입금 ===")
print(c.notes["차입금"])

# 전체 키 목록
print("\n영문 키:", c.notes.keys())
print("한글 키:", c.notes.keys_kr())
```

## 공시 텍스트

숫자 뒤에 숨겨진 텍스트도 분석한다.

```python
# 사업의 내용
sections = c.business
for s in sections:
    print(f"[{s.key}] {s.title} ({s.chars:,}자)")

# MD&A
print("\n=== MD&A ===")
print(c.mdna)

# 회사 개요
print("\n=== 회사 개요 ===")
ov = c.overview
if ov:
    print(f"설립: {ov.founded}")
    print(f"소재지: {ov.address}")
    print(f"종속회사: {ov.subsidiaryCount}개")
```

## 전체 일괄 조회

`all()`을 호출하면 모든 모듈을 순회하면서 전체 데이터를 dict로 반환한다.

```python
d = c.all()

print("전체 키:", list(d.keys()))
print()
print("=== 재무상태표 ===")
print(d["BS"])
print()
print("=== 배당 ===")
print(d["dividend"])
```

## Result 객체 접근

property는 대표 DataFrame 하나를 반환한다. 모듈의 전체 데이터가 필요하면 `get()`을 사용한다.

```python
# property → 대표 DataFrame
print("=== property (감사의견 DataFrame) ===")
print(c.audit)
print()

# get() → 전체 Result 객체
result = c.get("audit")
print("=== get() Result ===")
print("감사의견:", result.opinionDf)
print("감사보수:", result.feeDf)
```
