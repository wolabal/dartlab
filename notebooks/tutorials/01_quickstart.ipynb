## 4. Bridge Matching 상세

`fsSummary()` 메서드로 Bridge Matching 결과를 확인한다. 계정명 변경을 자동 추적한 통계를 볼 수 있다.

```python
result = samsung.fsSummary()

print(f"분석 연도: {result.nYears}년")
print(f"전체 매칭률: {result.allRate:.1%}")
print(f"변경점: {result.nBreakpoints}개")
print(f"연속 구간: {result.nSegments}개")
```

```python
# 연도별 매칭
for pair in result.pairResults:
    print(f"{pair.prevYear} → {pair.curYear}: {pair.rate:.1%} ({pair.matched}/{pair.total})")
```

```python
# 변경점
for bp in result.breakpoints:
    print(f"{bp.prevYear} → {bp.curYear}: {bp.rate:.1%}")
```

```python
# 연속 구간
for seg in result.segments:
    print(f"{seg.startYear}~{seg.endYear} ({seg.nYears}년, 매칭률 {seg.rate:.1%})")
```

## 5. 분기별·반기별 분석

`fsSummary()` 메서드의 `period` 파라미터로 분석 단위를 변경한다.

| 값 | 의미 | 포함 보고서 |
|-----|------|------------|
| `"y"` | 연간 (기본값) | 사업보고서 |
| `"q"` | 분기별 | 1분기 + 반기 + 3분기 + 사업 |
| `"h"` | 반기별 | 반기 + 사업 |

```python
# 분기별 재무상태표
quarterly = samsung.fsSummary(period="q")
quarterly.BS
```

```python
# 반기별 손익계산서
semiannual = samsung.fsSummary(period="h")
semiannual.IS
```
