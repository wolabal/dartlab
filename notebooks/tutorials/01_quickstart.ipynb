## 1. Company 생성

모든 분석의 시작점이다. 종목코드 또는 회사명으로 생성한다.

```python
from dartlab import Company

# 종목코드로 생성
samsung = Company("005930")
print(f"회사명: {samsung.corpName}")
print(f"종목코드: {samsung.stockCode}")

# 회사명으로도 생성 가능
kakao = Company("카카오")
print(f"\n회사명: {kakao.corpName}")
print(f"종목코드: {kakao.stockCode}")
```

## 2. 공시 목록 확인

해당 종목의 공시 목록과 DART 뷰어 링크를 조회한다.

```python
docs = samsung.docs()
docs
```

## 3. 재무제표 — property 접근

property 한 줄로 재무제표 DataFrame을 받는다. 괄호 없이 속성처럼 접근한다.

```python
print("=== 재무상태표 ===")
print(samsung.BS)
print("\n=== 손익계산서 ===")
print(samsung.IS)
print("\n=== 현금흐름표 ===")
print(samsung.CF)
```

## 4. Bridge Matching 상세

`fsSummary()` 메서드로 Bridge Matching 결과를 확인한다. 계정명 변경을 자동 추적한 통계를 볼 수 있다.

### Bridge Matching 결과 상세

연도별 매칭 결과, 변경점, 연속 구간을 확인한다.

```python
result = samsung.fsSummary()

print(f"분석 연도: {result.nYears}년")
print(f"전체 매칭률: {result.allRate:.1%}")
print(f"변경점: {result.nBreakpoints}개")
print(f"연속 구간: {result.nSegments}개")

print("\n--- 연도별 매칭 ---")
for pair in result.pairResults:
    print(f"{pair.prevYear} → {pair.curYear}: {pair.rate:.1%} ({pair.matched}/{pair.total})")

print("\n--- 변경점 ---")
for bp in result.breakpoints:
    print(f"{bp.prevYear} → {bp.curYear}: {bp.rate:.1%}")

print("\n--- 연속 구간 ---")
for seg in result.segments:
    print(f"{seg.startYear}~{seg.endYear} ({seg.nYears}년, 매칭률 {seg.rate:.1%})")
```

## 5. 분기별·반기별 분석

`fsSummary()` 메서드의 `period` 파라미터로 분석 단위를 변경한다.

| 값 | 의미 | 포함 보고서 |
|-----|------|------------|
| `"y"` | 연간 (기본값) | 사업보고서 |
| `"q"` | 분기별 | 1분기 + 반기 + 3분기 + 사업 |
| `"h"` | 반기별 | 반기 + 사업 |

```python
# 분기별
quarterly = samsung.fsSummary(period="q")
print("=== 분기별 재무상태표 ===")
print(quarterly.BS)

# 반기별
semiannual = samsung.fsSummary(period="h")
print("\n=== 반기별 손익계산서 ===")
print(semiannual.IS)
```

## 8. 전체 일괄 조회

`all()`을 호출하면 40개 모듈을 순회하면서 전체 데이터를 dict로 반환한다.

```python
d = samsung.all()

print("전체 키:", list(d.keys()))
print()
print("=== 재무상태표 ===")
print(d["BS"])
print()
print("=== 배당 ===")
print(d["dividend"])
```
