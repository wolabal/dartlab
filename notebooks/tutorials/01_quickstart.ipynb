## 1. Company 생성

모든 분석의 시작점이다. 종목코드 또는 회사명으로 생성한다.

```python
from dartlab import Company

samsung = Company("005930")
print(f"회사명: {samsung.corpName}")
print(f"종목코드: {samsung.stockCode}")
```

## 2. 공시 목록 확인

해당 종목의 공시 목록과 DART 뷰어 링크를 조회한다.

```python
docs = samsung.docs()
docs
```

## 3. 요약재무정보 분석

Bridge Matching으로 계정명 변경을 추적하면서 시계열을 생성한다.

```python
result = samsung.analyze()

print(f"분석 연도: {result.nYears}년")
print(f"전체 매칭률: {result.allRate:.1%}")
print(f"변경점: {result.nBreakpoints}개")
print(f"연속 구간: {result.nSegments}개")
```

```python
result.FS
```

### Bridge Matching 결과

연도별 매칭 결과와 변경점을 확인한다.

```python
for pair in result.pairResults:
    print(f"{pair.prevYear} → {pair.curYear}: {pair.rate:.1%} ({pair.matched}/{pair.total})")

print("\n--- 변경점 ---")
for bp in result.breakpoints:
    print(f"{bp.prevYear} → {bp.curYear}: {bp.rate:.1%}")

print("\n--- 연속 구간 ---")
for seg in result.segments:
    print(f"{seg.startYear}~{seg.endYear} ({seg.nYears}년, 매칭률 {seg.rate:.1%})")
```

## 4. 연결재무제표

재무상태표, 손익계산서, 현금흐름표를 상세 항목 수준으로 추출한다.

```python
stmt = samsung.statements()

print("=== 재무상태표 ===")
print(stmt.BS)
print("\n=== 손익계산서 ===")
print(stmt.IS)
print("\n=== 현금흐름표 ===")
print(stmt.CF)
```

## 5. 분기별 분석

`period` 파라미터로 분석 단위를 변경한다.

```python
quarterly = samsung.analyze(period="q")
quarterly.FS
```
